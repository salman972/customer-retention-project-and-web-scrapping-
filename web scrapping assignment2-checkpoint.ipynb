Q1: Write a python program to scrape data for “Data Analyst” Job position in 
“Bangalore” location. You have to scrape the job-title, job-location, company_name,
experience_required. You have to scrape first 10 jobs data.

This task will be done in following steps:
1. first get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill,Designations,Companies” field and enter “Bangalore” 
in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.
Note- All of the above steps have to be done in code. No step is to be done manually.

```python
import selenium
import pandas as pd
from selenium import webdriver
import time
```

```python
driver=webdriver.Chrome(r'chromedriver.exe')
driver
```

```python
url="https://www.naukri.com/"
driver.get(url)
```

```python
search_job_title=driver.find_element_by_id("qsb-keyword-sugg")
search_job_title.send_keys("Data Analyst")
```

```python
search_job_location=driver.find_element_by_id("qsb-location-sugg")
search_job_location.send_keys("Bangalore")
```

```python
search_button=driver.find_element_by_xpath("//button[@class='btn']")
search_button.click()
```

```python
titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
titles
```

```python
title=[]
for i in titles:
    title.append(i.text)
title
```

```python
companies=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
companies
```

```python
company=[]
for i in companies:
    company.append(i.text)
company
```

```python
experiences=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']")
experiences
```

```python
experience=[]
for i in experiences:
    experience.append(i.text)
experience
```

```python
locations=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
locations
```

```python
location=[]
for i in locations:
    location.append(i.text)
location
```

```python
df=pd.DataFrame({})
df['Title']=title
df['Company']=company
df['Experience,(in yrs)']=experience
df['Location']=location
df
```

```python
df=df.head(10)
df
```

```python
df['Experience,(in yrs)']=df['Experience,(in yrs)'].apply(lambda x:x[0:-4])
```

Q2: Write a python program to scrape data for “Data Scientist” Job position in
“Bangalore” location. You have to scrape the job-title, job-location,
company_name, full job-description. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. first get the webpage https://www.naukri.com/
2. Enter “Data Scientist” in “Skill,Designations,Companies” field and enter
“Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
import selenium
import pandas as pd
import time
from selenium import webdriver
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.naukri.com/")
```

```python
designation_search=driver.find_element_by_id("qsb-keyword-sugg")
designation_search.send_keys("Data Scientist")
location_search=driver.find_element_by_id("qsb-location-sugg")
location_search.send_keys("Bangalore")
search_btn=driver.find_element_by_xpath("//div[@class='search-btn']")
time.sleep(4)
search_btn.click()
```

```python
job_titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
title=[]
for i in job_titles:
    title.append(i.text)
title
```

```python
companies=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
company=[]
for i in companies:
    company.append(i.text)
company
```

```python
job_locations=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']")
location=[]
for i in job_locations:
    location.append(i.text)
location
```

```python
job_descriptions=driver.find_elements_by_xpath("//div[@class='job-description fs12 grey-text']")
job_description=[]
for i in job_descriptions:
    job_description.append(i.text)
job_description
```

```python
df=pd.DataFrame({})
df['Job_Title']=title
df['Job_Location']=location
df['Company']=company
df['Job Description']=job_description
df
```
