Q2: Write a python program to scrape data for “Data Scientist” Job position in
“Bangalore” location. You have to scrape the job-title, job-location,
company_name, full job-description. You have to scrape first 10 jobs data.
This task will be done in following steps:
1. first get the webpage https://www.naukri.com/
2. Enter “Data Scientist” in “Skill,Designations,Companies” field and enter
“Bangalore” in “enter the location” field.
3. Then click the search button.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data.

```python
import selenium
import pandas as pd
import time
from selenium import webdriver
```

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.naukri.com/")
```

```python
designation_search=driver.find_element_by_id("qsb-keyword-sugg")
designation_search.send_keys("Data Scientist")
location_search=driver.find_element_by_id("qsb-location-sugg")
location_search.send_keys("Bangalore")
search_btn=driver.find_element_by_xpath("//div[@class='search-btn']")
time.sleep(4)
search_btn.click()
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute("href"))
```

```python
job_title=[]
job_location=[]
company_name=[]
description=[]
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
for i in urls:
    driver.get(i)
    time.sleep(3)
    try:
        job=driver.find_element_by_xpath("//h1[@class='jd-header-title']")
        job_title.append(job.text)
    except NoSuchElementException:
        job_title.append('-')
    try:
        location=driver.find_element_by_xpath("//div[@class='loc']")
        job_location.append(location.text)
    except NoSuchElementException:
        job_location.append('-')
    try:
        company= driver.find_element_by_xpath("//a[@class='pad-rt-8']")
        company_name.append(company.text)
    except NoSuchElementException:
        company_name.append('-')
    try:
        des=driver.find_element_by_xpath("//section[@class='job-desc']")
        description.append(des.text)
    except NoSuchElementException:
        description.append('-')
```

```python
location=driver.find_element_by_xpath("//div[@class='loc']")
location.text
```

```python
df=pd.DataFrame({})
df['Job_Title']=job_title
df['Job_Location']=job_location
df['Company']=company_name
df['Job Description']=description
df1=df.drop([2,3,5,7,16,18],axis=0)
df1.reset_index(inplace=True,drop=True)
df1=df1.head(10)
df1
```
