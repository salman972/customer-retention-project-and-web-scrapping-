7. Write a python program to scrape details of all the mobile phones under Rs. 20,000 
listed on Amazon.in. The scraped data should include Product Name, Price, Image URL 
and Average Rating.

```python
import requests
from bs4 import BeautifulSoup
```

```python
page=requests.get('https://www.amazon.in/s?bbn=1389401031&rh=n%3A1389401031%2Cp_36%3A1318506031&dc&qid=1620204627&rnid=1318502031&ref=lp_1389401031_nr_p_36_3')
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
name2=soup.find_all('span',class_="a-size-base-plus a-color-base a-text-normal")
name2
```

```python
names=[]
for i in name2:
    names.append(i.text.replace('\n',''))
names
```

```python
price=soup.find_all('span',class_="a-price-whole")
price
```

```python
prices=[]
for i in price:
    prices.append(i.text.replace('\n',''))
prices
```

```python
ratings=soup.find_all("span",class_="a-declarative")
ratings
```

```python
rating=[]
for i in ratings:
    rating.append(i.text.replace('\n',''))
rating
```

```python
rating=rating[1:]
rating
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({})
df['Name']=names
df['Price']=prices
df['Rating']=rating
df
```
