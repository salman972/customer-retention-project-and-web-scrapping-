1. Scrape the details of most viewed videos on YouTube from Wikipedia:
Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
 You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views


```python
#importing libraries
import selenium
from selenium import webdriver
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")#taking chrome to wikipedia site
```

```python
driver.execute_script("window.scrollBy(0,1000)")
```

```python
ranks=[]
names=[]
artists=[]
views=[]
dates=[]
```

```python
for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[1]"):
        ranks.append(i.text.replace('.',''))
for j in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[2]"):
        names.append(j.text)
for k in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[3]"):
        artists.append(k.text)
for l in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[4]"):
        views.append(l.text)
for m in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']/tbody/tr/td[5]"):
        dates.append(m.text)
```

```python
# making dataframe
import pandas as pd
df=pd.DataFrame({})
df['ranks']=ranks[0:30]
df['names']=names[0:30]
df['artists']=artists[0:30]
df['views(in billions)']=views[0:30]
df['dates']=dates[0:30]
df
```

```python
df['names']=df['names'].apply(lambda x:x.split('[')[0])#removing bracket and its insides
df["names"]=df["names"].str.replace('"','')#removing "" from name column
df
```

5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/
You have to find the following details:
A) Repository title
B) Repository description
C) Contributors count
D) Language used

```python
#importing necessary libraries
import selenium
from selenium import webdriver
driver=webdriver.Chrome(r'chromedriver.exe')
driver.get(" https://www.github.com/")#load url
import time
```

```python
explore=driver.find_element_by_xpath("//html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[1]/li/a").get_attribute("href")
driver.get(explore)#take url to explore
```

```python
driver.find_element_by_xpath("//div[@class='d-flex flex-wrap flex-items-center flex-justify-center flex-md-justify-start text-center text-md-left']/a[3]").click()
```

```python
urls=[]
for i in driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a"):
    urls.append(i.get_attribute("href"))
```

```python
title=[]
description=[]
count=[]
used=[]
count1=[]
count2=[]
```

```python
from selenium.common.exceptions import NoSuchElementException
```

```python
for i in urls:
    driver.get(i)
    time.sleep(4)
    try:
        title.append(driver.find_element_by_xpath("//h1[@class=' d-flex flex-wrap flex-items-center break-word f3 text-normal']").text.replace('\n',''))
    except NoSuchElementException:
        title.append('--')
    try:
        description.append(driver.find_element_by_xpath("//p[@class='f4 mt-3']").text)
    except NoSuchElementException:
        description.append('--')
    try:
        count.append(driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[5]/div/h2/a").text)
    except NoSuchElementException:
        count.append('--')
    try:
        count1.append(driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[4]/div/h2/a").text)
    except NoSuchElementException:
        count1.append('--')
    try:
        count2.append(driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[3]/div/h2/a").text)
    except NoSuchElementException:
        count2.append('--')
        
    try:
        used.append(driver.find_element_by_xpath("//span[@class='color-text-primary text-bold mr-1']").text)
    except NoSuchElementException:
        used.append('--')
    
    
```

```python
#making dataframe
import pandas as pd
df=pd.DataFrame({})
df['title']=title
df['description']=description
df['used']=used
df['count']=count
df['count1']=count1
df['count2']=count2
df
```

```python
df=df.replace("Packages",'--')#replace Packages by --
df['count1'].loc[7]="--"#replace used by numbers with --
df['count2'].loc[[9,23]]="--"#replace used by numbers with --
```

```python
df['Count']=df['count']+df['count1']+df['count2']
import re
df['Count']=df['Count'].apply(lambda x:re.sub('[^0-9]','',x))#extract only numbers
df=df.drop(['count','count1','count2'],axis=1)
df
```
